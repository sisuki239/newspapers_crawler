```python
import pandas as pd
```

```python
# open "vnexpress_articles.csv" file 
df = pd.read_csv("vnexpress_articles.csv")
# check if the file is loaded correctly
if df.empty:
    print("The file is empty or not loaded correctly.")
```

```python
df
```

```python
# search for articles url to take decimal numbers that has length of 7, apply regex. For all row, make this number the article id
df['article_id'] = df['url'].str.extract(r'(\d{7})')
# check if the article_id column is created correctly
if df['article_id'].isnull().all():
    print("No article IDs were extracted. Please check the URL format.")
```

```python
df
```

```python
# save the updated DataFrame to a new CSV file
df.to_csv("vnexpress_articles_with_ids.csv", index=False)
```

```python
df2 = pd.read_csv("vnexpress_articles_with_ids.csv")
df3 = pd.read_csv("./output/vnx_truy_quet_buon_lau_hang_gia_filtered.csv")
```

```python
# find the union of the two DataFrames based on the 'article_id' column
df_union = pd.merge(df2, df3, on='article_id', how='outer')
```

```python
df_union
```

```python
# join url_x and url_y columns into a new column named 'url'
df_union['url'] = df_union['url_x'].combine_first(df_union['url_y'])
# move all "headline" to "title" column
df_union['title'] = df_union['headline'].combine_first(df_union['title'])
# join description_x and description_y columns into a new column named 'description'
df_union['description'] = df_union['description_x'].combine_first(df_union['description_y'])
# drop the old columns that are no longer needed
df_union = df_union.drop(columns=['url_x', 'url_y', 'headline', 'description_x', 'description_y'])
```

```python
df_union.to_csv("vnexpress_articles_union.csv", index=False)
```
